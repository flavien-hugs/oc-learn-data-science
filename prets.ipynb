```python
# https://colab.research.google.com/github/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/blob/main/notebooks/P2/P2C2%20-%20Manipulez%20un%20dataframe.ipynb#scrollTo=193e1033

import pandas as pd
import numpy as np

prets = pd.read_csv("data/prets.csv")
prets.head()
```

```python
# cr√©ez une nouvelle variable calculant le taux d'endettement de chaque individu.
# Ce taux correspond au pourcentage du revenu rembours√© chaque mois par un individu.
# Vous arrondirez le r√©sultat √† 2 chiffres apr√®s la virgule.

taux_endettement = (prets["remboursement"] / prets["taux"]) * 100
prets["taux_endettement"] = round(taux_endettement, 2)
prets.head()
```

```python
# renommer la variable taux en taux_interet pour eviter toute confusion:
prets.rename(columns={"taux": "taux_interet"}, inplace=True)
prets.head()
```

```python
"""
Nous allons √† pr√©sent cr√©er deux derni√®res variables :

- 'cout_total' correspondant au cout total du pret √† partir du remboursement et de la duree
- 'benefices' correspondant aux b√©n√©fices mensuels r√©alis√©s par la banque sur le pr√™t

Nous simplifierons ici le calcul des b√©n√©fices par :

b√©n√©fices =  (ùê∂xùëá) 24 

avec:

C = cout total du pr√™t
T = taux d'int√©r√™t
"""

prets["cout_total"] = prets["remboursement"] * prets["duree"]

benefice_mensuel_realise = ((prets['cout_total'] * prets['taux_interet']) / 100) / 24
prets["benefices"] = round(benefice_mensuel_realise, 2)

prets.head()
```

```python
# connaitre les 5 pr√™ts sur lesquels nous produisons le plus de b√©n√©fices. Mettez en place une solution pour les afficher :
prets.sort_values(by="benefices", ascending=False)
prets.head()
```

```python
# Le taux d‚Äôendettement autoris√© est de 35%. Pourriez vous me communiquer le nombre de personnes ayant d√©pass√© ce seuil ?

mask_endettement_autorise = prets['taux_endettement'] > 35
clients_risque = prets.loc[mask_endettement_autorise, :]
clients_risque.shape[0]
```

```python
# M√™me question, mais cette fois ci uniquement sur l‚Äôagence Parisienne
client_risque_paris = prets.loc[(mask_endettement_autorise) & (prets["ville"] == "PARIS"), :]
client_risque_paris.shape[0]
```

```python
"""
Pour faciliter le traitement d‚Äô√©ventuelles futures demandes de pr√™ts, pourriez vous ajouter une variable nomm√©e 
risque qui aurait pour valeur Oui si le client est risqu√© (taux d‚Äôendettement > 35%), Non sinon.
Pour se faire, je vous conseille de cr√©er la variable risque en l'initialisant √† Non (ou Oui, √† votre pr√©f√©rence) et de ne modifier
que les lignes concern√©es par la valeur inverse :
"""

prets['risque'] = 'Non'
prets.loc[mask_endettement_autorise, 'risque'] = 'Oui'
prets
```

```python
# Combien de pr√™ts automobiles ont √©t√© accord√©s ? Quel est le co√ªt total moyen de ces derniers ?
prets_auto = prets.loc[prets["type"] == "automobile", :]

nombre_prets_auto = prets_auto.shape[0]
cout_total_moyen = prets_auto["cout_total"].mean()

print('Nous avons accord√©', nombre_prets_auto, 'pr√™ts automobiles, dont le co√ªt total moyen est de', round(cout_total_moyen, 2), '‚Ç¨')
```

```python
# Quel est le b√©n√©fice mensuel total r√©alis√© par l‚Äôagence Toulousaine ?
benefice_mensuel_agence_toulouse = prets.loc[prets["ville"] == "TOULOUSE", "benefices"].sum()
round(benefice_mensuel_agence_toulouse, 2)
```

```python
# calculer l‚Äôinformation recherch√©e initialement, soit le chiffre d'affaires total de chacune de nos agences ?

prets.groupby("ville").sum()
```

```python
# par agence ET par type de pr√™t
prets.groupby(["ville", "type"]).sum()
```

```python
# sur la variable remboursement
prets.groupby(["ville", "type"])["remboursement"].sum()
```

```python
# calcule la moyenne et la somme de remboursement par agence, ainsi que le maximum de revenu
remb_agence = prets.groupby("ville").agg({"remboursement": ["sum", "mean"], "revenu": "max"})
remb_agence
```

```python
# Agr√©gez des lignes et des colonnes: Calcule du chiffre d'affaire d√©gag√© par chaque agence en fonction du type de pr√™t contracter
ca_agence = prets.pivot_table(index="ville", columns="type", values="remboursement", aggfunc="sum")
ca_agence
```

```python
# cr√©er un data frame de profil client pour avoir par client toutes les informations qui le concernent, r√©sum√©es en une ligne
profile_clients = prets.groupby("identifiant")[['revenu','remboursement','duree','taux_endettement','cout_total','benefices']].sum()
profile_clients.reset_index(inplace=True)
profile_clients.head()
```

```python
# calculer le nombre exact de personnes en √©tat de situation bancaire risqu√©e, √† partir du taux d‚Äôendettement et de ces profils clients
nombre_client_risque = profile_clients.loc[profile_clients["taux_endettement"] > 35, :].shape[0]
print('Il y a', nombre_client_risque, 'clients qui ont d√©pass√© le seuil autoris√©')
```

```python
# calculer le b√©n√©fice total d√©gag√© par chacune de nos agences, par type de pr√™t
benefice_total_par_agence = prets.groupby(['ville', 'type'])["benefices"].sum()
benefice_total_par_agence.reset_index()
```

```python
# calculer les b√©n√©fices moyens r√©alis√©s par chaque agence, pour chaque type de pr√™t, sous forme de tableau √† double entr√©e 
benefices_moyens = prets.pivot_table(index="ville", columns="type", values="benefices", aggfunc="mean")
benefices_moyens.reset_index()
```
