```python
# https://colab.research.google.com/github/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/blob/main/notebooks/P2/P2C2%20-%20Manipulez%20un%20dataframe.ipynb#scrollTo=193e1033

import pandas as pd
import numpy as np

prets = pd.read_csv("data/prets.csv")
prets.head()
```

```python
# créez une nouvelle variable calculant le taux d'endettement de chaque individu.
# Ce taux correspond au pourcentage du revenu remboursé chaque mois par un individu.
# Vous arrondirez le résultat à 2 chiffres après la virgule.

taux_endettement = (prets["remboursement"] / prets["taux"]) * 100
prets["taux_endettement"] = round(taux_endettement, 2)
prets.head()
```

```python
# renommer la variable taux en taux_interet pour eviter toute confusion:
prets.rename(columns={"taux": "taux_interet"}, inplace=True)
prets.head()
```

```python
"""
Nous allons à présent créer deux dernières variables :

- 'cout_total' correspondant au cout total du pret à partir du remboursement et de la duree
- 'benefices' correspondant aux bénéfices mensuels réalisés par la banque sur le prêt

Nous simplifierons ici le calcul des bénéfices par :

bénéfices =  (𝐶x𝑇) 24 

avec:

C = cout total du prêt
T = taux d'intérêt
"""

prets["cout_total"] = prets["remboursement"] * prets["duree"]

benefice_mensuel_realise = ((prets['cout_total'] * prets['taux_interet']) / 100) / 24
prets["benefices"] = round(benefice_mensuel_realise, 2)

prets.head()
```

```python
# connaitre les 5 prêts sur lesquels nous produisons le plus de bénéfices. Mettez en place une solution pour les afficher :
prets.sort_values(by="benefices", ascending=False)
prets.head()
```
