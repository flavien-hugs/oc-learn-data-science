```python
"""
Appliquez la sélection via des indices

Le premier type de sélection est réalisé via des indices. Il se fait via la méthode .iloc

un indice est la position à laquelle se trouve un élément au sein d’un tableau, que ce soit une liste,
une Series ou un array. En Python, les indices commencent à 0 : ainsi, le premier élément aura pour indice 0, le second 1, etc.

La méthode '.iloc' suit la syntaxe suivante: mon_dataframe.iloc[indice_ligne, indice_colonne]
"""

import pandas as pd

clients = pd.read_csv("data/P2C1_CLIENTS/clients.csv")
clients
```

```python
# sélectionner le nom du premier client de notre base de données
clients.iloc[0,2]
```

```python
# Comme avec les arrays, on peut utiliser l’opérateur ':' pour sélectionner la plage souhaitée

# sélectionner le nom des 10 premiers clients :
clients.iloc[:10, 2]
```

```python
# sélectionner toutes les colonnes des 10 premiers clients :
clients.iloc[:10, :]
```

```python
# sélectionner l’email et le nom des 10 clients suivants:
clients.iloc[10:20, 1:3]
```

```python
# sélectionner l’e-mail et le genre des 10 derniers clients :
clients.iloc[-10:, [1, 3]]
```

```python
# la sélection via des conditions: (.loc | syntaxe: mon_dataframe.loc[ condition sur les lignes, colonne(s) ]

import pandas as pd

prets = pd.read_csv("data/prets.csv")
prets.head()
```

```python
# lister les personnes ayant contracter un pret de type automobile ET un revenu superieur ou egal a 4000$

prets.loc[(prets["type"] == "automobile") & (prets["revenu"] >= 4000), :]
```

```python
# lister les prets de type automobile 

prets.loc[prets["type"] == "automobile", :]
```

```python
# lister les prets de type automobile ainsi que les identifiants des personnes

prets.loc[prets["type"] == "automobile", "identifiant"]
```

```python
# lister les prets de type automobile ainsi que les identifiants et leurs revenus

prets.loc[prets["type"] == "automobile", ["identifiant", "revenu"]]
```

```python
# lister les personnes ayant contracter un pret de type automobile, qui sont dans la ville de paris et qui ont un revenu
# superieur ou egal 4000$

prets.loc[(prets["type"] == "automobile") & (prets["ville"] == "PARIS") & (prets["revenu"] >= 4000), :]
```
